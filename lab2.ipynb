```python
# Завдання 7
# Імпортуємо бібліотеку Pandas
import pandas as pd

# Завантажуємо дані
df = pd.read_csv('NationalNames.csv')

# Додаємо стовпець 'Id', який є індексом рядка
df['Id'] = df.index + 1

# Фільтруємо дані: вибираємо чоловічі імена у 2010 році
df_male_2010 = df[(df['Gender'] == 'M') & (df['Year'] == 2010)]

# Групуємо за іменами та підраховуємо їх кількість
top_5_male_names_2010 = df_male_2010.groupby(['Id', 'Name', 'Year', 'Gender'], as_index=False)['Count'].sum()

# Сортуємо за кількістю і вибираємо топ-5 імен
top_5_male_names_2010 = top_5_male_names_2010.sort_values(by='Count', ascending=False).head(5)

# Виводимо результат у вигляді таблиці
print(top_5_male_names_2010.to_string(index=False))
```

```python
# завдання 18
# Імпортуємо бібліотеку Pandas
import pandas as pd

# Завантажуємо дані
df = pd.read_csv('NationalNames.csv')

# Групуємо за роком та статтю, підсумовуємо кількість народжень
births_by_gender_per_year = df.groupby(['Year', 'Gender'])['Count'].sum().unstack()

# Порівнюємо кількість народжених дівчаток та хлопчиків для кожного року
more_girls_than_boys = births_by_gender_per_year['F'] > births_by_gender_per_year['M']

# Підраховуємо кількість років, коли дівчаток народжувалось більше
count_years_more_girls = more_girls_than_boys.sum()

print(count_years_more_girls)
```

```python
# завдання 23
import pandas as pd

# Завантажуємо набір даних
df = pd.read_csv('NationalNames.csv')

# Фільтруємо імена, які зустрічаються кожного року (у всіх роках)
names_per_year = df.groupby(['Name', 'Year'])['Count'].sum().reset_index()

# Тепер знаходимо імена, які є в кожному році
names_per_year_count = names_per_year.groupby('Name').size()

# Фільтруємо тільки ті імена, що зустрічаються в кожному році
names_in_all_years = names_per_year_count[names_per_year_count == len(df['Year'].unique())].index.tolist()

# Створюємо список популярних імен
popular_names = df[df['Name'].isin(names_in_all_years)].groupby('Name')['Count'].sum().reset_index()
popular_names = popular_names.sort_values(by='Count', ascending=False)

# Виводимо 15 найпопулярніших гендерно-нейтральних імен
top_15_names = popular_names.head(15)

# Виводимо результат без індексів
print(top_15_names[['Name']].to_string(index=False))
```

```python
# завдання 24
import pandas as pd

# Завантажуємо дані
df = pd.read_csv('NationalNames.csv')

# Визначаємо непопулярні імена (з малим кількістю використань)
low_popularity_names = df[df['Count'] <= df['Count'].quantile(0.1)]  # Нижні 10%

# Знаходимо найпопулярніше ім'я серед непопулярних
most_popular_low_name = low_popularity_names.groupby('Name')['Count'].sum().idxmax()

# Знаходимо кількість використань цього імені
most_popular_low_name_count = low_popularity_names.groupby('Name')['Count'].sum().max()

# Вивести найпопулярніше непопулярне ім'я і кількість його використань
print(f"Найпопулярніше непопулярне ім'я: {most_popular_low_name} (Кількість використань: {most_popular_low_name_count})")
```

```python
# завдання 25
import pandas as pd
import matplotlib.pyplot as plt

# Завантажуємо дані
df = pd.read_csv('NationalNames.csv')

# Фільтруємо дані для імен John та Mary
names_to_filter = ['John', 'Mary']
df_filtered = df[df['Name'].isin(names_to_filter)]

# Групуємо дані по роках та іменах, і сумуємо кількість використань
df_grouped = df_filtered.groupby(['Year', 'Name'])['Count'].sum().reset_index()

# Створюємо графік
plt.figure(figsize=(10, 6))

# Побудова графіку для імені John
plt.plot(df_grouped[df_grouped['Name'] == 'John']['Year'],
         df_grouped[df_grouped['Name'] == 'John']['Count'], label='John', color='blue')

# Побудова графіку для імені Mary
plt.plot(df_grouped[df_grouped['Name'] == 'Mary']['Year'],
         df_grouped[df_grouped['Name'] == 'Mary']['Count'], label='Mary', color='red')

# Налаштування графіку
plt.title('Розподіл кількості імен John та Mary по роках')
plt.xlabel('Рік')
plt.ylabel('Кількість використань')
plt.legend()
plt.grid(True)

# Показати графік
plt.show()
```

```python
# завдання 26
import pandas as pd
import matplotlib.pyplot as plt

# Завантажуємо дані
df = pd.read_csv('NationalNames.csv')

# Фільтруємо дані для жіночого імені John та чоловічого імені Mary
female_john = df[(df['Name'] == 'John') & (df['Gender'] == 'F')]
male_mary = df[(df['Name'] == 'Mary') & (df['Gender'] == 'M')]

# Групуємо дані по роках для кожного імені та сумуємо кількість використань
female_john_distribution = female_john.groupby('Year')['Count'].sum()
male_mary_distribution = male_mary.groupby('Year')['Count'].sum()

# Створюємо графік
plt.figure(figsize=(10, 6))

# Побудова графіку для жіночого імені John
plt.plot(female_john_distribution.index, female_john_distribution, label='Жіноче ім’я John', color='red')

# Побудова графіку для чоловічого імені Mary
plt.plot(male_mary_distribution.index, male_mary_distribution, label='Чоловіче ім’я Mary', color='blue')

# Налаштування графіку
plt.title("Розподіл кількості жіночих імен John та чоловічих імен Mary по роках")
plt.xlabel("Рік")
plt.ylabel("Кількість народжень")
plt.legend()
plt.grid(True)

# Показати графік
plt.show()

```

```python
# завдання 27
import pandas as pd

# Завантажуємо набір даних
df = pd.read_csv('NationalNames.csv')

# Групуємо за роками та іменами, знаходимо максимальну кількість для кожного року
most_popular_names = df.groupby(['Year', 'Name'])['Count'].sum().reset_index()

# Для кожного року вибираємо найпопулярніше ім'я
most_popular_names_per_year = most_popular_names.loc[most_popular_names.groupby('Year')['Count'].idxmax()]

# Виводимо таблицю з роками, найпопулярнішими іменами та кількістю без індексів
result = most_popular_names_per_year[['Year', 'Name', 'Count']]

# Виведення результату без індексів
print(result.to_string(index=False))

# Збереження результату в CSV файл без індексів
result.to_csv('most_popular_names_per_year_with_count.csv', index=False)
```
